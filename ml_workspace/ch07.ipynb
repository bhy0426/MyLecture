```python
import numpy as np
from sklearn.linear_model import SGDRegressor
from sklearn.utils import shuffle

X = np.random.uniform(0, 2, 100).reshape(-1, 1)# 0~2 사이 난수
noise = np.random.normal(0, 30, 100) # 평균0 표준편차1 노이즈

#3x+4+노이즈
y = 3 * X[:,0] + 4 + noise

# 3. SGD 모델 생성
model = SGDRegressor(
    loss="squared_error",
    learning_rate="constant",
    shuffle=True)

# 4. 학습
model.fit(X, y)
```

```python
y_hat = model.predict(X)
y_hat[:5]
```

```python
import pandas as pd

#데이터 가져오기
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data'
data = pd.read_csv(data_url, sep='\s+', header = None) #정규 표현식:'\s+' 공백 하나 이상

#데이터의 열 이름 지정
data.columns = ['CRIM','ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO' ,'B', 'LSTAT', 'MEDV']
data.head()
```

```python
y = data.pop('MEDV')
```

```python
y = y.values
x = data.values
```

```python
from sklearn.preprocessing import MinMaxScaler

# 객체 생성
mms = MinMaxScaler(feature_range=(0,5)) # feature_range 벙뮈 설정

mms1 = mms.fit(x) # 계산하기 위한 기본 값 생성
reX = mms1.transform(x) # 범위에 맞게 실제로 x 값을 변경
# reX = mms.fis_transform(x) # 각 열의 min, max를 구하고 값을 변경해줌, 위 두 줄을 한 번에 실행
reX[:5]
```

```python
from sklearn.model_selection import train_test_split

trX, teX, trY, teY = train_test_split(reX, y, test_size=0.33)
print(trX.shape)
print(teX.shape)
print(trY.shape)
print(teY.shape)
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression(fit_intercept=True,copy_X=True, n_jobs=8)
SGD_regr = linear_model.SGDRegressor(penalty="l2", alpha=0.01, max_iter=1000, tol=0.001, eta0=0.01)
```

```python
# 학습(가중치를 구함)
regr.fit(trX, trY)
SGD_regr.fit(trX, trY)
```

```python
print(regr.coef_)
print(SGD_regr.coef_)
```

```python
print(regr.intercept_)
print(SGD_regr.intercept_)
```

```python
y1 = regr.predict(teX)
y2 = SGD_regr.predict(teX)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

r2_score(y1, teY), r2_score(y2, teY)
```

```python
mean_squared_error(y1, teY), mean_squared_error(y2, teY)
```

```python
mean_absolute_error(y1, teY), mean_absolute_error(y2, teY)
```
