```python
# 202204013 변하연
import pandas as pd

path1 = "https://raw.githubusercontent.com/prof-dana/Lecture-Code/refs/heads/main/MachineLearning/data/HW/GlobalHumanDataset.csv" 
path2 = "https://raw.githubusercontent.com/prof-dana/Lecture-Code/refs/heads/main/MachineLearning/data/HW/LifeExpectancy.csv"

df_ghd = pd.read_csv(path1)
df_le = pd.read_csv(path2)

print(df_ghd.head(5))
print(df_le.head().T)
```

```python
df_mean_height = df_ghd.groupby("country")["height"].mean().reset_index()
df_mean_weight = df_ghd.groupby("country")["weight"].mean().reset_index()
print(df_mean_height.head())
print(df_mean_weight.head())
```

```python
df_mean_inner = pd.merge(left=df_mean_height, right=df_mean_weight, how="inner", on="country")
print(df_mean_inner.head())
```

```python
df_mean_inner["bmi"] = ((df_mean_inner["weight"]) / (df_mean_inner["height"]*0.01)**2)
print(df_mean_inner.head())
```

```python
cols= ["lifeExp","gdp","alc","edu","adultMort","incomeComp","totExp"]
df_le.dropna(subset=cols, inplace=True)
print(df_le.isnull().sum() / len(df_le))
```

```python
df_le.fillna(df_mean_inner["bmi"][df_mean_inner["country"] == "Sudan"].values[0], inplace=True)
df_le[df_le["country"] == "Sudan"]
print(df_le.isnull().sum() / len(df_le))
```

```python
df_le.drop("country", axis=1, inplace=True)
df_le = pd.get_dummies(df_le)
print(df_le.head(3).T)
```

```python
from sklearn.model_selection import train_test_split

Y = df_le.lifeExp.values.reshape(-1, 1)
df_le.drop("lifeExp", axis=True,inplace=True)
X = df_le.values

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

regr = LinearRegression()
regr.fit(X_train, Y_train)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import r2_score

Y_hat = regr.predict(X_test)

print('MSE:', mean_squared_error(Y_test, Y_hat))
print('MAE:', mean_absolute_error(Y_test, Y_hat))
print('RMSE:', root_mean_squared_error(Y_test, Y_hat))
print('R2:', r2_score(Y_test, Y_hat))
```
