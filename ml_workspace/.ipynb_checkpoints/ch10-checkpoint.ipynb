```python
import pandas as pd

df = pd.read_csv("./data/UVA_Internet_Survey_Dataset.csv")
df.head()
```

```python
df.drop(["no", "who", "Years on Internet"], axis=1, inplace=True)
```

```python
df.drop("Country", axis=1, inplace=True)
```

```python
df.dtypes
```

```python
df
```

```python
category_cols = ['Gender', 'Household Income', 
                 'Sexual Preference','Education Attainment', 
                 'Major Occupation', "Marital Status"]

for col in category_cols:
    df[col] = df[col].astype('category')#범주형 데이터 타입
 
df.dtypes
```

```python
df["Household Income"]
```

```python
df2 = pd.get_dummies(df)
df2.head().T
```

```python
df2.isnull().sum()
```

```python
df2.fillna({"Age": df2["Age"].mean()}, inplace=True)
```

```python
df2.isnull().sum()
```

```python
y = df2.pop("Newbie")
x = df2
```

```python
from sklearn.model_selection import train_test_split

# 먼저 훈련, 테스트 데이터를 분리하고 정규화 진행하기
trX, teX, trY, teY = train_test_split(x, y) # 기본값은 25% 비율로 테스트 데이터 분리
trX.shape
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

ss = ss.fit(trX) # trX 평균, 표준편차 계산

trX2 = ss.transform(trX) # 실제 계산
trX2[:5]
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
```

```python
lr.fit(trX2, trY) # 가중치를 계산해서 lr에 저장

# 시그모이드 함수 1/1+e^-z <- z= w0+ w1*Age(1) + ...
```

```python
teX.head()
```

```python
# 꼭 기존에 사용했던 StandardScaler객체로 정규화를 진행하기
teX2 = ss.transform(teX) # trX의 평균, 표준편차를 가지고 teX의 정규화를 진행
```

```python
teX2[:4]
```

```python
y2 = lr.predict(teX2) # 예측값
```
