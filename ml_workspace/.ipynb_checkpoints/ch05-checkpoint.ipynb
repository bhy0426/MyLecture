```python
import pandas as pd
import numpy as np

df = pd.DataFrame(
    {'A':[14.00,90.20,90.95,96.27,91.21],
     'B':[103.02,107.26,110.35,114.23,114.68], 
     'C':['big','small','big','small','small']})

df
```

```python
Amin = df['A'].min() # 시리즈 객체(넘파이 배열 벡터)
Amax = df['A'].max() # 시리즈 객체(넘파이 배열 벡터)

nor = (df['A'] - Amin) / (Amax - Amin)

# AMin, AMax # 괄호가 생략된 튜플 출력
nor
```

```python
Bmean = df["B"].mean() # 평균
Bstd = df["B"].std() # 표준편차

z =  (df['B'] - Bmean) / Bstd

z
```

```python
ctg_columns = ["Pclass", "Gender", "Embarked"]

for c in ctg_columns:
    onehot = pd.get_dummies(df[c], prefix=c)#해당 열 원핫인코딩
    
    #index로 병합하여 다시 df에 저장 df = pd.concat([df, onehot], axis=1)
    df = pd.merge(df,onehot, how="inner", left_index=True, right_index=True)

df.head(2).T
```
