```python
import matplotlib.pyplot as plt
import numpy as np

m = 100

X = np.random.uniform(0, 100, m)
Y = 3*X+ 30 + np.random.normal(0, 30, 100)

# 그래프
plt.scatter(X, Y, color="red")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# 객체 생성, 아무 일도 일어나지 않음, 가중치를 찾기 위한 준비
m = LinearRegression()

reX = X.reshape(-1, 1) # 열 1개인 2차원 벡터

# 실제로 가중치를 찾아나섬 => 학습, 최적의 가중치를 찾음
# 최소자승법 사용
m.fit(reX, Y)

print(m.coef_) # 열에 대한 가중치
print(m.intercept_) # 상수값
```

```python
y_hat = m.predict(reX) # 2차원 배열로 받음
y_hat
```

```python
plt.scatter(X, Y, color="red")
plt.plot(X,y_hat)
plt.show()
```

```python
bias = np.ones((100, 1))
X = np.hstack((reX, bias)) # 세로로 결합

theta = np.random.uniform(0, 1, 2) # w1, w0

h = np.dot(X, theta) # 예측값
loss = h - Y # 오차
cost = (np.sum(loss ** 2 ) / (2 * 100)) # 비용함수
gradient = np.dot(X.T, loss)/100 # 0으로 근접하도록
theta = theta * 0.005 * gradient

theta
```
